```python
import os
import cv2
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
uid_key = "sub-101"
slice_num = 30
original_pixel = "/mnt/storage/ji/brain_mri_valdo_mayo/valdo_resample_ALFA_YOLO_PNG_epd_gt_box_t2s/images/train"
mean_pixel = "/mnt/storage/ji/brain_mri_valdo_mayo/valdo_resample_ALFA_YOLO_PNG_epd_gt_box_t2s/images/train_mean_pixel"
```

```python
slice_filename_original = f"{uid_key}_slice_{slice_num:03d}.png"
slice_path_original = os.path.join(original_pixel, slice_filename_original)
image_original = Image.open(slice_path_original)
plt.imshow(image_original, cmap='gray')
```

```python
image_array_original = np.array(image_original)
print(image_array_original.shape)  # Prints the dimensions of the image
```

### Mean Pixel Intensity

```python
slice_filename_mean_pixel = f"{uid_key}_slice_{slice_num:03d}.png"
slice_path_mean_pixel = os.path.join(mean_pixel, slice_filename_mean_pixel)
image_mean_pixel = Image.open(slice_path_mean_pixel)

image_array_mean_pixel = np.array(image_mean_pixel)
print(image_array_mean_pixel.shape)  # Prints the dimensions of the image
plt.imshow(image_array_mean_pixel, cmap='gray')
```

```python
image_array_mean_pixel = np.array(image_array_mean_pixel)
print(image_array_mean_pixel.shape)  # Prints the dimensions of the image
```

### Comparison

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_original, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(image_array_mean_pixel, cmap='gray')
plt.title("Mean Pixel Image")
plt.axis("off")
```

```python
if np.array_equal(image_array_original, image_array_mean_pixel):
    print("Same")
else:
    print("Not same")
```

```python
difference = np.abs(image_array_original - image_array_mean_pixel)

# Print statistics about the difference
print(f"Maximum difference: {np.max(difference)}")
print(f"Minimum difference: {np.min(difference)}")
print(f"Mean difference: {np.mean(difference)}")
print(f"Standard deviation of difference: {np.std(difference)}")

# Optionally, visualize the difference as an image (if applicable)
plt.imshow(difference, cmap='hot')
plt.colorbar()
plt.title("Difference Heatmap")
plt.show()
```
