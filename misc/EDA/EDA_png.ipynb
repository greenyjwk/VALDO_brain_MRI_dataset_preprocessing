```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
uid = "322"
slice = "075"
patch_id = "0"
```

### ground truth box

```python
# Load the image
image_path = f"/mnt/storage/ji/brain_mri_valdo_mayo/YOLO_valdo_stacked_1mm_png_temp_gt_boxes/sub-{uid}_slice_{slice}_boxes.png"
image = Image.open(image_path)

# Convert the image to an ndarray
image_array = np.array(image)

# Separate the channels
red_channel = image_array[:, :, 0]
green_channel = image_array[:, :, 1]
blue_channel = image_array[:, :, 2]

# Plot the channels
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(red_channel, cmap='gray')
plt.title('T1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(green_channel, cmap='gray')
plt.title('T2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(blue_channel, cmap='gray')
plt.title('T2S')
plt.axis('off')

plt.show()
```
