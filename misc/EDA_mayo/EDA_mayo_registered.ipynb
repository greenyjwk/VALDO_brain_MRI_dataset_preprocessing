```python
import os
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
```

```python
MRN = "4817777"
slice = 28
```

```python
root = "/media/Datacenter_storage/Ji/brain_mri_valdo_mayo/mayo_registered_resampled_0325_TEMP"

T1 = nib.load(os.path.join(root, MRN, f"3D_SAG_T1_MPRAGE_1MM_{MRN}.nii.gz"))
T1 = T1.get_fdata()

T2 = nib.load(os.path.join(root, MRN, f"SAG_3D_T2_SPACE_{MRN}.nii.gz"))
T2 = T2.get_fdata()

T2S = nib.load(os.path.join(root, MRN, f"Axial_3TE_T2_STAR_{MRN}.nii.gz"))
T2S = T2S.get_fdata()
```

```python
print(T2S.shape)

plt.figure(figsize=(10,5))

# Plot histograms
plt.figure(figsize=(10,5))
plt.hist(T2S[:,:,slice].ravel(), bins=100, alpha=0.5, label='Before correction')
# plt.hist(].ravel(), bins=100, alpha=0.5, label='After correction')
plt.xlabel('Intensity')
plt.ylabel('Voxel count')
plt.title('Intensity Histogram Before vs. After Bias Field Correction')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(131)
plt.imshow(np.rot90(T1[:,:,slice], k=-1), cmap='gray', origin='lower')
plt.title('T1')

plt.subplot(132)
plt.imshow(np.rot90(T2[:,:,slice], k=-1), cmap='gray', origin='lower')
plt.title('T2')

plt.subplot(133)
plt.imshow(np.rot90(T2S[:,:,slice, 2], k=-1), cmap='gray', origin='lower')
plt.title('T2S')
```
