```python
import os
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
MRN = "10745491"
slice = 30
```

```python
root = "/media/Datacenter_storage/Ji/brain_mri_valdo_mayo/mayo_stacked"

img = nib.load(os.path.join(root, MRN, f"{MRN}.nii.gz"))
img = img.get_fdata()
print(img.shape)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(131)
plt.imshow(img[:,:,slice, 0], cmap='gray', origin='lower')
plt.title('T1')

plt.subplot(132)
plt.imshow(img[:,:,slice, 1], cmap='gray', origin='lower')
plt.title('T2')

plt.subplot(133)
plt.imshow(img[:,:,slice, 2], cmap='gray', origin='lower')
plt.title('T2S')
```
