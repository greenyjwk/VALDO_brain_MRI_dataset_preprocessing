```python
import os
import matplotlib.pyplot as plt
import nibabel as nib
```

```python
path = "/media/Datacenter_storage/Ji/brain_mri_valdo_mayo/mayo_resampled/7800567"

img = [None] * 3
for root, _, files in os.walk(path):
    print(root)
    print(files)

for files in files:
    if files.startswith('3D'):
        img[0] = os.path.join(root, files)
    elif files.startswith('Axial'):
        img[2] = os.path.join(root, files)
    elif files.startswith('SAG'):
        img[1] = os.path.join(root, files)
```

```python
slice = 40

img0 = nib.load(img[0])
img0 = img0.get_fdata()

img1 = nib.load(img[1])
img1 = img1.get_fdata()

img2 = nib.load(img[2])
img2 = img2.get_fdata()

print(img0.shape)
print(img1.shape)
print(img2.shape)
```

```python
plt.figure(figsize=(15, 15))

plt.subplot(1,3,1)
plt.imshow(img0[:,:,slice], cmap='gray')

plt.subplot(1,3,2)
plt.imshow(img1[:,:,slice], cmap='gray')

plt.subplot(1,3,3)
plt.imshow(img2[:,:,slice, 2], cmap='gray')
```
