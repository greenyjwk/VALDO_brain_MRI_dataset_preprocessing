```python
import sys
import subprocess
import os

root_path = "/media/Datacenter_storage/Ji/brain_mri_valdo_mayo/mayo_skull_stripped_resampled_0325"
sequence = "Axial_3TE_T2_STAR"

for file in os.listdir(root_path):
    file_path = os.path.join(root_path, file, f"{sequence}_{file}.nii.gz")
    print(file_path)

    # Define the output directory
    output_dir = f"/media/Datacenter_storage/Ji/csf_segment/{file}/T2"

    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        print(f"Creating output directory: {output_dir}")
        os.makedirs(output_dir)
        print("Directory created successfully.")
    else:
        print("Output directory already exists.")

    # Define the command to run
    command = [
        "fast", 
        "-t", "2", 
        "-n", "3", 
        "-g", 
        "-o", output_dir,
        f"{file_path}"
    ]

    # Run the command
    try:
        print("Running fast command...")
        result = subprocess.run(command, check=True, capture_output=True, text=True)
        print("Command completed successfully.")
        print("Output:", result.stdout)
    except subprocess.CalledProcessError as e:
        print(f"Command failed with exit code {e.returncode}")
        print("Error output:", e.stderr)
```
