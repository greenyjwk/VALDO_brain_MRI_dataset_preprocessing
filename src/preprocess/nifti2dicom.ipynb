```python
output_dir = "/media/Datacenter_storage/valdo_dicom/Task2"
root_dir = "/media/Datacenter_storage/valdo_dataset/Task2"
for sub_dir in os.listdir(root_dir):
    temp_output_dir =""
    sub_dir_root = os.path.join(root_dir, sub_dir)
    nii_file = glob.glob(os.path.join(sub_dir_root, "*T2S.nii.gz"))
    if len(nii_file) == 0:
        continue
    nii_file = nii_file[0]
    if not os.path.exists(os.path.join(output_dir, sub_dir)):
        temp_output_dir = os.path.join(output_dir, sub_dir)
        os.makedirs(temp_output_dir, exist_ok = True)

        print(nii_file)
        print(temp_output_dir)
        print()

        cmd = [
            'nifti2dicom',
            '-i', nii_file,
            '-o', output_dir
        ]

        result = subprocess.run(cmd, capture_output=True, text=True)
```
